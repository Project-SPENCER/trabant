```python
import os
import itertools
import glob
import datetime

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.ticker
import matplotlib.pyplot as plt
```

```python
image_log = pd.read_csv("image_log.csv")
bupt_trajectory_sl = pd.read_csv("bupt_trajectory_sl.csv")
```

```python
image_log.head()
```

```python
bupt_trajectory_sl.head()
```

```python
# merge bupt_trajectory_sl on image_log with time_ms_since_1682379292000 on t
new_df = image_log.merge(bupt_trajectory_sl, left_on="t", right_on="time_ms_since_1682379292000")
new_df["t_ms"] = new_df["t"]
new_df["lat"] = new_df["lat_x"]
new_df["lon"] = new_df["lon_x"]
new_df["alt"] = new_df["alt_km"]
new_df["is_sunlit"] = new_df["is_sunlit"].apply(lambda x: 1 if x == 1.0 else 0)
new_df[["t_ms", "lat", "lon", "alt", "is_sunlit"]].head()
```

```python
new_df[["t_ms", "lat", "lon", "alt", "is_sunlit"]].to_csv("image_log_with_alt.csv", index=False)
```
